## Estadisticas descriptivas

```python
plot_names = ['Estado de cuenta en SEP', 'Estado de cuenta en AGO', 'Estado de cuenta en JUL',
              'Estado de cuenta en JUN', 'Estado de cuenta en MAY', 'Estado de cuenta en ABR',
              'Cantidad pagada en SEP', 'Cantidad pagada en AGO', 'Cantidad pagada en JUL',
              'Cantidad pagada en JUN', 'Cantidad pagada en MAY', 'Cantidad pagada en ABR']


plt.figure(figsize=(15, 8))
for i, column in enumerate(df.columns[11:23], start=1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=df[column], color='skyblue')
    plt.title(plot_names[i-1])

plt.tight_layout()
plt.show()
```

## Graficos de barra para variables categoricas 

```python
# Definir etiquetas para las variables categóricas
gender_labels = {1: 'Masculino', 2: 'Femenino'}
education_labels = {1: 'Posgrado', 2: 'Universidad', 3: 'Bachillerato', 4: 'Otro'}
marital_status_labels = {1: 'Casado', 2: 'Soltero', 3: 'Otro'}


# Crear un diccionario de etiquetas para cada variable categórica
labels_dict = {'X2': gender_labels, 'X3': education_labels, 'X4': marital_status_labels}

# Establecer el estilo de los gráficos
plt.style.use('ggplot')

# Crear subplots para cada variable categórica
fig, axs = plt.subplots(1, len(categorical_columns), figsize=(15, 5))

# Iterar sobre las columnas categóricas y graficar la frecuencia de los datos con las etiquetas
for i, column in enumerate(categorical_columns):
    labels = labels_dict[column]
    df[column].map(labels).value_counts().plot(kind='bar', ax=axs[i], color='skyblue')
    axs[i].set_title(f'Distribución de {column}', fontsize=14)
    axs[i].set_xlabel(column, fontsize=12)
    axs[i].set_ylabel('Frecuencia', fontsize=12)
    axs[i].tick_params(axis='x', labelrotation=45)

plt.tight_layout()
plt.show()
```

```python
payment_status_labels = {-1: 'Pago puntual', 1: 'Retraso 1 mes', 2: 'Retraso 2 meses', 3: 'Retraso 3 meses', 
                         4: 'Retraso 4 meses', 5: 'Retraso 5 meses', 6: 'Retraso 6 meses', 7: 'Retraso 7 meses', 
                         8: 'Retraso 8 meses', 9: 'Retraso mas de 9 meses'}

labels_dict = {'X6': payment_status_labels, 'X7': payment_status_labels, 'X8': payment_status_labels,
               'X9': payment_status_labels, 'X10': payment_status_labels, 'X11': payment_status_labels}

plot_names = ['Estado de pago en SEP', 'Estado de pago en AGO', 'Estado de pago en JUL', 
              'Estado de pago en JUN', 'Estado de pago en MAY', 'Estado de pago en ABR']

plt.style.use('ggplot')

for i, (column, labels) in enumerate(labels_dict.items()):
    fig, ax = plt.subplots(figsize=(6, 4))
    df[column].map(labels).value_counts().plot(kind='bar', color='skyblue')
    ax.set_title(f'Distribución de {plot_names[i]}', fontsize=14)
    ax.set_xlabel(column, fontsize=12)
    ax.set_ylabel('Frecuencia', fontsize=12)
    ax.tick_params(axis='x', labelrotation=45)
    plt.tight_layout()
    plt.show() 
```

```python
df_continuas = df[['X1','X5','X12','X13','X14','X15','X16','X17','X18','X19','X20','X21','X22','X23','Y']]
corr = df_continuas.corr()
plt.figure(figsize=(20, 20)) 
sns.heatmap(corr, cmap='coolwarm', annot=True, xticklabels=corr.columns,
            yticklabels=corr.columns)
plt.title('Matriz de correlación')
plt.xticks(rotation=45, ha="right") 
plt.show()
```
