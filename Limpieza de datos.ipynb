# Limpieza de datos

```python
import pandas as pd
```

#### Llamar a la base de datos

```python

# Cargar los datos
from ucimlrepo import fetch_ucirepo
default_of_credit_card_clients = fetch_ucirepo(id=350)
X = default_of_credit_card_clients.data.features
y = default_of_credit_card_clients.data.targets

# Convertir X e y a DataFrames
X_df = pd.DataFrame(data=X)
y_df = pd.DataFrame(data=y)

# Asignar nombres de columna genéricos a X
X_df.columns = [f"X{i}" for i in range(1, X_df.shape[1] + 1)]

# Combinar X e y en un único DataFrame
df = pd.concat([X_df, y_df], axis=1)

# Verificar los primeros registros del DataFrame
print(df.head())
df
```

### Reemplzar los valores de 0 por -1

```python
# Reemplazar los valores de 0 con -1 en las columnas X6 hasta X11
df.loc[:, 'X6':'X11'] = df.loc[:, 'X6':'X11'].replace(0, -1)

# Verificar que los valores hayan sido reemplazados correctamente
print(df[['X6', 'X7', 'X8', 'X9', 'X10', 'X11']].head())
```

### Reemplazar los valores de 5 y 6 en la columna de X3

```python
df.loc[:, 'X3'] = df.loc[:, 'X3'].replace(5, 4)
df.loc[:, 'X3'] = df.loc[:, 'X3'].replace(6, 4)
#Verificar los valores que tiene ahora
print("Valores únicos de la variable X3:", df['X3'].unique())
```
